```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#laoad dataset
cafe = pd.read_csv('dirty_cafe_sales.csv')
cafe.head(10)
```

```python
cafe.describe()
```

```python
cafe.info()
```

```python
cafe[['Quantity', 'Price Per Unit', 'Total Spent']] = (cafe[['Quantity', 'Price Per Unit', 'Total Spent']].replace({
    'ERROR':np.nan,
    'UNKNOWN':np.nan
})
)
cafe.head(10)
    
```

```python
cafe[['Item', 'Payment Method', 'Location']] = (cafe[['Item', 'Payment Method', 'Location']].replace({
    'ERROR':'Unknown'
})
)
cafe.head(10)
    
```

```python
cafe[['Item', 'Payment Method', 'Location']] = cafe[['Item', 'Payment Method', 'Location']].apply(lambda col:col.str.title())
cafe.head(10)
```

```python
cafe['Quantity'] = pd.to_numeric(cafe['Quantity'], errors='coerce')
cafe['Price Per Unit'] = pd.to_numeric(cafe['Price Per Unit'], errors='coerce')
cafe['Total Spent'] = pd.to_numeric(cafe['Total Spent'], errors='coerce')
cafe.head()
```

```python
#identify row for review
mask = cafe['Quantity'] * cafe['Price Per Unit'] != cafe['Total Spent']
cafe[mask]
```

```python
cafe.loc[mask, 'Total Spent'] = cafe.loc[mask, 'Quantity'] * cafe.loc[mask, 'Price Per Unit']
cafe.head(10)
```

```python
cafe.info()
```

```python
cafe.nunique()
```

```python
cafe['Item'].unique()
```

```python
cafe['Payment Method'].unique()
```

```python
cafe['Location'].unique()
```

```python
cafe['Transaction Date'] = pd.to_datetime(cafe['Transaction Date'], errors='coerce')
cafe.head(20)
```

```python
cafe.info()
```

```python
cafe.dtypes
```

```python
#Top Selling Items
top_items = cafe['Item'].value_counts()
top_items.head(10)
```

```python
#Total Revenue
total_revenue = cafe['Total Spent'].sum()
print("Total Revenue:", total_revenue)
```

```python
#Revenue by item category
total_revenue = cafe['Total Spent'].sum()
print("Total Revenue:", total_revenue)
```

```python
#Revenue By location
revenue_by_location = cafe.groupby('Location')['Total Spent'].sum().sort_values(ascending=False)
print(revenue_by_location)
```

```python
#Most common payment method
payment_method_counts = cafe['Payment Method'].value_counts()
print(payment_method_counts)
```

```python
#Average spend per transaction
avg_spend = cafe['Total Spent'].mean()
print("Average Spend per Transaction:", avg_spend)
```

```python
#Monthly Sales Trend
monthly_sales = cafe.groupby(cafe['Transaction Date'].dt.to_period('M'))['Total Spent'].sum()
print(monthly_sales)
```

```python
#Busiest Months and lowest months
busiest_month = monthly_sales.idxmax(), monthly_sales.max()
slowest_month = monthly_sales.idxmin(), monthly_sales.min()
print("Busiest Month:", busiest_month)
print("Slowest Month:", slowest_month)
```

```python
#Day of week sales performance
day_of_week_sales = cafe.groupby(cafe['Transaction Date'].dt.day_name())['Total Spent'].sum()
print(day_of_week_sales)
```

```python
plt.figure(figsize=(8,5))
top_items.head(5).plot(kind="bar", color="skyblue")
plt.title("Top 5 Items by Sales Volume")
plt.xlabel("Item")
plt.ylabel("Quantity Sold")
plt.xticks(rotation=45)
plt.show()
```

```python
payment_counts = cafe['Payment Method'].value_counts()
plt.figure(figsize=(6,4))
payment_counts.plot(kind="bar", color="green")
plt.title("Payment Method Distribution")
plt.xlabel("Payment Method")
plt.ylabel("Number of Transactions")
plt.show()
```

```python
plt.figure(figsize=(10,5))
monthly_sales.plot(color="black", marker="o")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(cafe['Total Spent'], bins=30, color='yellow', edgecolor="black")
plt.title("Distribution of Transaction Amounts")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
location_revenue = cafe.groupby('Location')['Total Spent'].sum()

plt.figure(figsize=(8,6))
location_revenue.plot(kind="bar", color="Orange")
plt.title("Revenue by Location")
plt.xlabel("Location")
plt.ylabel("Total Revenue")
plt.show()
```
